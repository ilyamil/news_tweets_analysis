```python
import mlflow
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
from news_tweets_analysis.preprocessing import TextPreprocessor

ROOT = Path().resolve().parent.as_posix()
TRACKING_URI = f'file://{ROOT}/mlruns'
EXPERIMENT = 'classic-models'
```

```python
mlflow.set_tracking_uri(TRACKING_URI)
if not mlflow.get_experiment_by_name(EXPERIMENT):
    mlflow.create_experiment(EXPERIMENT)
mlflow.set_experiment(EXPERIMENT)
```

```python
runs = mlflow.search_runs(
    experiment_names=[EXPERIMENT],
    filter_string='attribute.status = "FINISHED"',
    order_by=['metrics.val_score DESC']
)
# runs[runs['tags.mlflow.runName']]
#runs['tags.mlflow.runName']
#runs.groupby('tags.mlflow.note.content')
runs
```
